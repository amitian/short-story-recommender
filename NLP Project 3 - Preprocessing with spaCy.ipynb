```python
#Spacy
import pandas as pd
import numpy as np
import spacy
sp = spacy.load('en_core_web_lg')

pd.set_option('max_colwidth',200)
corpus_df = pd.read_pickle('corpus_df3_cleaned.pkl')
corpus_df.head()
```

```python
# Previewing nouns, verbs, entities
story = sp(corpus_df.TEXT[0])
print("Noun phrases:", [chunk.text for chunk in story.noun_chunks][0:25])
print("Verbs:", [token.lemma_ for token in story if token.pos_ == "VERB"][0:25])

# Find named entities, phrases and concepts
for entity in story.ents[0:20]:
    print(entity.text, entity.label_)
```

```python
#finding nouns from text
def noun_finder(docs):
    nouns = []
    text = sp(docs)
    for word in text:
        if word.pos_ == 'NOUN':
            nouns.append(word.lemma_)
    return nouns

corpus_df['TEXT_NOUNS'] = corpus_df.TEXT.map(lambda x: noun_finder(x))

corpus_df.head()
```

```python
#nouns + adjectives
def noun_adj_finder(docs):
    nouns_adjs = []
    text = sp(docs)
    for word in text:
        if (word.pos_ == 'NOUN') | (word.pos_ == 'ADJ'):
            nouns_adjs.append(word.lemma_)
    return nouns_adjs

corpus_df['TEXT_NOUNS_ADJS'] = corpus_df.TEXT.map(lambda x: noun_adj_finder(x))

corpus_df.head()
```

```python
#nouns + verbs
def noun_verb_finder(docs):
    nouns_verbs = []
    text = sp(docs)
    for word in text:
        if (word.pos_ == 'NOUN') | (word.pos_ == 'VERB'):
            nouns_verbs.append(word.lemma_)
    return nouns_verbs

corpus_df['TEXT_NOUNS_VERBS'] = corpus_df.TEXT.map(lambda x: noun_verb_finder(x))

corpus_df.head()
```

```python
#nouns + verbs + adjectives
def noun_verb_adj_finder(docs):
    nouns_verbs_adjs = []
    text = sp(docs)
    for word in text:
        if (word.pos_ == 'NOUN') | (word.pos_ == 'VERB') | (word.pos_ == 'ADJ'):
            nouns_verbs_adjs.append(word.lemma_)
    return nouns_verbs_adjs

corpus_df['TEXT_NOUNS_VERBS_ADJS'] = corpus_df.TEXT.map(lambda x: noun_verb_adj_finder(x))

corpus_df.head()
```

```python
stopwords = sp.Defaults.stop_words

def remove_stopwords(text):
    words = []
    for token in text:
        if token.lower() not in stopwords:
            words.append(token)
    return words

corpus_df['TEXT_NOUNS'] = corpus_df.TEXT_NOUNS.map(lambda x: remove_stopwords(x))
corpus_df['TEXT_NOUNS_ADJS'] = corpus_df.TEXT_NOUNS_ADJS.map(lambda x: remove_stopwords(x))
corpus_df['TEXT_NOUNS_VERBS'] = corpus_df.TEXT_NOUNS_VERBS.map(lambda x: remove_stopwords(x))
corpus_df['TEXT_NOUNS_VERBS_ADJS'] = corpus_df.TEXT_NOUNS_VERBS_ADJS.map(lambda x: remove_stopwords(x))

corpus_df.head()
```

```python
def find_ents(docs):
    words = []
    text = sp(docs)
    for ent in text.ents:
        if (ent.label_ == 'GPE') | (ent.label_ == 'LOC'):
            words.append(ent.text)
    return words

corpus_df['TEXT_ENTS'] = corpus_df.TEXT.map(lambda x: find_ents(x))

corpus_df.head()
```

```python
corpus_df = pd.read_pickle("corpus_df_spacy3.pkl")
```

```python
#remove stopwords in TEXT_ENTS
corpus_df['TEXT_ENTS'] = corpus_df.TEXT_ENTS.map(lambda x: remove_stopwords(x))

#turning list into string
corpus_df['TEXT_NOUNS'] = corpus_df.TEXT_NOUNS.map(lambda x: ' '.join (str(e) for e in x))
corpus_df['TEXT_NOUNS_ADJS'] = corpus_df.TEXT_NOUNS_ADJS.map(lambda x: ' '.join (str(e) for e in x))
corpus_df['TEXT_NOUNS_VERBS'] = corpus_df.TEXT_NOUNS_VERBS.map(lambda x: ' '.join (str(e) for e in x))
corpus_df['TEXT_NOUNS_VERBS_ADJS'] = corpus_df.TEXT_NOUNS_VERBS_ADJS.map(lambda x: ' '.join (str(e) for e in x))

corpus_df.head()
```

```python
#second round of cleaning
import re
import string

corpus_df['TEXT_NOUNS'] = corpus_df.TEXT_NOUNS.map(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
corpus_df['TEXT_NOUNS'] = corpus_df.TEXT_NOUNS.map(lambda x: re.sub('[‘’“”…:.,?()_♦]', '', x))
corpus_df['TEXT_NOUNS_VERBS'] = corpus_df.TEXT_NOUNS_VERBS.map(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
corpus_df['TEXT_NOUNS_VERBS_ADJS'] = corpus_df.TEXT_NOUNS_VERBS_ADJS.map(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
corpus_df['TEXT_NOUNS_VERBS'] = corpus_df.TEXT_NOUNS_VERBS.map(lambda x: re.sub('[‘’“”…:.,?()_♦]', '', x))
corpus_df['TEXT_NOUNS_VERBS_ADJS'] = corpus_df.TEXT_NOUNS_VERBS_ADJS.map(lambda x: re.sub('[‘’“”…:.,?()_♦]', '', x))
corpus_df.head()
```

```python
#removing \xa0 from original text
import re
corpus_df['TEXT'] = corpus_df.TEXT.map(lambda x: re.sub('xa0', '', x))
corpus_df['TEXT'] = corpus_df.TEXT.map(lambda x: re.sub('[\\\\]', '', x))
corpus_df.head()
```
